    <h3>Description:</h3>
    <p style="font-family: Times New Roman; font-size: 16px"><b>Abstract:</b> The dataset contains a subset of locations and attributes of incidents reported in the ASAP (Analytical Services Application) crime report database by the District of Columbia Metropolitan Police Department (MPD).
    <br>
    <br>
    Visit <a href="https://crimecards.dc.gov">https://crimecards.dc.gov</a> for more information. This data is shared via an automated process where addresses are geocoded to the District's Master Address Repository and assigned to the appropriate street block. Block locations for some crime points could not be automatically assigned resulting in 0,0 for x,y coordinates. These can be interactively assigned using the MAR Geocoder.
    <br>
    <br>
    On February 1 2020, the methodology of geography assignments of crime data was modified to increase accuracy. From October 20 2023 going forward, all crime data will have Ward, ANC, SMD, BID, Neighborhood Cluster, Voting Precinct, Block Group and Census Tract values calculated prior to, rather than after, anonymization to the block level. This change impacts approximately one percent of Ward assignments.</p>

```python
import pandas as pd
import numpy as np
import altair as alt
from summarytools import dfSummary
from datetime import datetime, timedelta
import pytz
```

```python
def convert_date(date_string):
    # Remove the timezone offset from the string
    date_string_without_tz = date_string[:-4]

    # Convert the string to datetime object
    date_object = datetime.strptime(date_string_without_tz, "%Y/%m/%d %H:%M:%S")

    # Convert the datetime object to desired date format
    formatted_date = date_object.strftime("%m-%d-%Y")
    
    return formatted_date
```

```python
df = pd.read_csv("Crime_Incidents_in_2023.csv")
df.head(3)
```

```python
dfSummary(df)
```

```python
df['REPORT_DAT'] = df['REPORT_DAT'].apply(convert_date)
df['REPORT_DAT'] = df['REPORT_DAT'].apply(pd.to_datetime, format="%m-%d-%Y")
df.head(2)
```

<p style="font-family: Times New Roman; font-size: 16px">On observing closely, we notice that there are variablesa that are not needed for our data visualization project. Hence we can eliminate variables such as: BLOCK, XBLOCK, YBLOCK, ANC, PSA, NEIGHBORHOOD_CLUSTER, BLOCK_GROUP, CENSUS_TRACT, VOTING_PRECINCT, BID, OBJECTID and OCTO_RECORD_ID.</p>

```python
df = df.drop(['BLOCK','XBLOCK','YBLOCK','ANC','PSA','NEIGHBORHOOD_CLUSTER','BLOCK_GROUP','CENSUS_TRACT',
              'VOTING_PRECINCT','BID','OBJECTID','OCTO_RECORD_ID'],axis=1)
dfSummary(df)
```

```python
df['DISTRICT'] = df['DISTRICT'].fillna(0.)
df.DISTRICT.unique()
```

```python
df['START_DATE'] = df['START_DATE'].apply(pd.to_datetime)
dfSummary(df)
```
